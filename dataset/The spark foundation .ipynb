# Predict the percentage of an student based on the no. of study hours.

step 1:Importing Python library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

step 2: Importing Data-set 

```python
data = pd.read_csv("data.csv")
print("Importing Data Successfully")
```

step 3:Check weather Data Imported Successfully or not?

```python
print ("For this we print first 10 data set of Data-set")
data.head(10)
```

```python
print("You have correctly Imported Data-set")
```

step 4:Plot the graph , for detail Analysis of Data-set

```python

data.plot(x='Hours',y='Scores',style='1')
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

```python
data.plot.pie(x='Hours',y='Scores')
```

```python
data.plot.scatter(x='Hours',y='Scores')
```

```python
data.plot.bar(x='Hours',y='Scores')
```

```python
data.sort_values(["Hours"],axis=0,
                 ascending=[True],inplace=True)
data.head(10)
data.plot.bar(x='Hours',y='Scores')
```

step 5: Now, we have prepared the data for our model

```python
X=data.iloc[:,:-1].values
y=data.iloc[:,1].values
#print(x)
```

step 6: Now, have divide the data for training & testing the model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    test_size=0.2, random_state=0)
```

Training The Algorithm

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

regressor.fit(X_train, y_train)
print("Training Complete.")
```

step 7: Now, our model is ready. Its time to test it.

```python
print(X_test)
print("Predection of Score")
y_pred = regressor.predict(X_test)
print(y_pred)
```

step 8: Now, Checking the accuracy of our model

```python
df = pd.DataFrame({'Actual': y_test,'Predicted': y_pred})
df
```

step 9: Now, its time to predict with custom input

```python
hours = [[9.25]]
pred = regressor.predict(hours)
print(pred)
```

# Evaluating the Model

```python
from sklearn import metrics
print ('Mean Absolute Error:',
      metrics.mean_absolute_error(y_test, y_pred))
```
